### Biblioteca Altair

Altair é uma poderosa biblioteca Python voltada para a visualização de dados estatísticos. Sua abordagem é declarativa, o que significa que você descreve o que deseja visualizar, e a biblioteca cuida de como a visualização será gerada. Essa característica simplifica o processo de criação de gráficos complexos, pois a sintaxe é intuitiva e se concentra em expressar a intenção do usuário.

Altair é baseada nas gramáticas de visualização Vega e Vega-Lite, que fornecem a estrutura subjacente para criar visualizações ricas e interativas. Essas gramáticas permitem que a aparência visual e o comportamento interativo de uma visualização sejam descritos em formato JSON, facilitando a personalização e a flexibilidade na construção de gráficos. Vega é mais completa e flexível, enquanto Vega-Lite é uma versão simplificada e mais acessível, ideal para a maioria dos casos de uso.

Altair se destaca por sua integração perfeita com bibliotecas de ciência de dados em Python, como Pandas, permitindo a criação de visualizações diretamente a partir de DataFrames. Além disso, a biblioteca suporta interatividade, tornando-a uma excelente escolha para explorar e comunicar insights de dados.

- APIs Imperativas
> Com APIs imperativas, você está explicitamente instruindo o computador sobre como executar uma tarefa. Isso significa que você deve especificar cada passo do processo de criação da visualização. Por exemplo, ao usar bibliotecas como Matplotlib, você precisa definir manualmente vários aspectos da visualização, como os limites dos eixos, a posição das legendas, o tamanho das fontes, entre outros detalhes. O foco está na execução passo a passo do código para criar o gráfico, o que pode exigir mais tempo e esforço, especialmente para visualizações complexas.

- APIs Declarativas
>  Por outro lado, as APIs declarativas, como o Altair, permitem que você se concentre em o que deseja visualizar. Em vez de se preocupar com cada detalhe de como o gráfico será construído, você simplesmente descreve a visualização em termos dos dados e da saída desejada. O Altair, então, cuida do restante, gerando automaticamente as melhores configurações para apresentar os dados de forma clara e informativa.

Vantagens do Altair
O Altair se destaca precisamente por essa natureza declarativa. Ele abstrai a complexidade de criação de gráficos e permite que você se concentre mais na análise dos dados do que na codificação dos gráficos. Isso é particularmente útil para cientistas de dados, que podem dedicar mais tempo para explorar, entender e interpretar os dados, sem precisar se preocupar com as minúcias de como os gráficos são construídos. O Altair também facilita a criação de visualizações interativas, tornando-o uma ferramenta poderosa para comunicar insights de forma eficaz.

O tutorial utilizará o Movies Dataset do Vega para realizar análises de dados e criar visualizações com Altair. O conjunto de dados inclui informações sobre filmes, como título, arrecadação, orçamento, gênero e classificações do IMDB e Rotten Tomatoes. O foco será em algumas colunas específicas para aprender como usar o Altair para visualização de dados, sem explorar todas as colunas do dataset.

```python
from vega_datasets import data as vega_data

filmes_df = pd.read_json(vega_data.movies.url)

print("filmes_df tem o seguinte tipo: ", type(filmes_df))

print("filmes_df é do seguinte tamanho: ", filmes_df.shape)
```

o *filmes_df* é na verdade um DataFrame pandas e tem 3201 filmes (linha) com 16 informações de recursos (colunas) cada. Vamos verificar os dados dentro do dataframe.

```python
filmes_df.head(5)
```

```python
filmes_df.columns
```

a função **to_datetime** dentro do pandas é usada para converter a coluna Release_Date em uma data real e então extrair o valor 'year' dela.

```python
def extract_year(value):
    return pd.to_datetime(value, format='%b %d %Y').year

filmes_df["Year"] = filmes_df["Release_Date"].apply(extract_year)
```

Veja que criamos uma nova coluna chamada YEAR

#### Gráficos

Charts são um objeto fundamental no Altair, e ele aceita um único argumento - um dataframe. A maneira como definimos um chart é a seguinte:

```python
alt.Chart(DataFrame)
```

##### Mark

A propriedade mark permite que você especifique como os dados precisam ser representados no gráfico. A seguir estão as propriedades básicas de mark fornecidas pelo Altair:

image.png

Para a propriedade mark, gostaríamos que cada valor no dataFrame fosse marcado como um ponto (círculo) no gráfico.

```python
alt.Chart(DataFrame).mark_point()
```

#### Codificações

Uma vez que temos os dados e como eles são representados - queremos especificar onde representá-los. Ou seja, configurar as posições, tamanho, cor, etc.

```python
alt.Chart(DataFrame).mark_point().encode()
```

Juntando tudo isso e adicionando detalhes

```python
chart = alt.Chart(filmes_2000).mark_point().encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross')
)
```

```python
display(chart)
```

```python
alt.Chart(filmes_2000).mark_point(filled=True).encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross'),
    alt.Size('US_Gross')
)
```

 Daremos a cada gênero do filme sua própria cor usando o recurso Color e adicionaremos alguma transparência à cor com OpacityValue .

```python
alt.Chart(filmes_2000).mark_point(filled=True).encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross'),
    alt.Size('US_Gross'),
    alt.Color('Major_Genre'),
    alt.OpacityValue(0.7)
)
```

Agora conseguiremos ver detalhes sobre o filme passando o mouse por cima do ponto de dados que representa.

```python
alt.Chart(filmes_2000).mark_point(filled=True).encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross'),
    alt.Size('US_Gross'),
    alt.Color('Major_Genre'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('Title'),
               alt.Tooltip('Production_Budget'),
               alt.Tooltip('Worldwide_Gross'),
               alt.Tooltip('US_Gross')
              ]
)
```

E se fizermos uma interação com o usuário? vamos adicionar uma funcionalidade simples:  interactive()

```python
alt.Chart(filmes_2000).mark_point(filled=True).encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross'),
    alt.Size('US_Gross'),
    alt.Color('Major_Genre'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('Title'),
               alt.Tooltip('Production_Budget'),
               alt.Tooltip('Worldwide_Gross'),
               alt.Tooltip('US_Gross')
              ]
).interactive()
```

E se a gente selecionar somente por ano?
 Além disso, definitivamente queremos separá-los com base no valor Year que criamos anteriormente.

```python
select_year = alt.selection_single(
    name='Select', fields=['Year'], init={'Year': 1928},
    bind=alt.binding_range(min=1928, max=2046, step=10)
)

alt.Chart(filmes_df).mark_point(filled=True).encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross'),
    alt.Size('US_Gross'),
    alt.Color('Major_Genre'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('Title:N'),
               alt.Tooltip('Production_Budget:Q'),
               alt.Tooltip('Worldwide_Gross:Q'),
               alt.Tooltip('US_Gross:Q')
              ]
).add_selection(select_year).transform_filter(select_year)
```

```python
print("O filme mais velho é de: ", filmes_df["Year"].min())
print("O filme mais novo é de: ", filmes_df["Year"].max())
```

Como você pode ver, ao arrastar a barra 'Select_Year', não há muitos filmes representados no ano de 1928 a 1967. E não há dados de filmes registrados depois de 2018 e certamente não até 2046. Podemos usar a propriedade binding_range() para especificar o intervalo nos anos e também os tamanhos dos passos dos anos.

```python
select_year = alt.selection_single(
    name='Select', fields=['Year'], init={'Year': 1968},
    bind=alt.binding_range(min=1968, max=2008, step=10)
)

alt.Chart(filmes_df).mark_point(filled=True).encode(
    alt.X('Production_Budget'),
    alt.Y('Worldwide_Gross'),
    alt.Size('US_Gross'),
    alt.Color('Major_Genre'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('Title:N'),
               alt.Tooltip('Production_Budget:Q'),
               alt.Tooltip('Worldwide_Gross:Q'),
               alt.Tooltip('US_Gross:Q')
              ]
).add_selection(select_year).transform_filter(select_year)
```
