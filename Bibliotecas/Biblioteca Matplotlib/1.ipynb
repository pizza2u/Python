#### $\color{red}{\text{O que é Matplotlib?}}$

Matplotlib é uma biblioteca de plotagem de gráficos de baixo nível em python que serve como um utilitário de visualização.

Matplotlib foi criado por John D. Hunter.

Matplotlib é de código aberto e podemos usá-lo livremente.

O Matplotlib é escrito principalmente em python, alguns segmentos são escritos em C, Objective-C e Javascript para compatibilidade com a plataforma.

#### $\color{red}{\text{Base de código do Matplotlib}}$
O código-fonte do Matplotlib está localizado neste [repositório](https://github.com/matplotlib/matplotlib) do github.

### $\color{blue}{\text{Introdução do Matplotlib}}$

#### $\color{red}{\text{Instalação}}$

Se você já tem o Python e o PIP instalados em um sistema, a instalação do Matplotlib é muito fácil.

Instale-o usando este comando:

> C:\Users\Your Name>pip install matplotlib

Se este comando falhar, é necessário usar uma distribuição python que já tenha o Matplotlib instalado, como Anaconda, Spyder etc.

#### $\color{red}{\text{Importar Matplotlib}}$

Depois que o Matplotlib estiver instalado, importe-o em seus aplicativos adicionando a instrução: `import module`

> import matplotlib

#### $\color{red}{\text{Versão}}$

A string da versão é armazenada no atributo `__version__` .



```python
import matplotlib

print(matplotlib.__version__)
```

### $\color{blue}{\text{Matplotlib Pyplot}}$


#### $\color{red}{\text{Pyplot}}$

A maioria dos utilitários do Matplotlib está no submódulo pyplot e geralmente são importados sob o alias **plt**:

> Para quem não lembra o que é alias: Na programação python, o segundo nome dado a um dado é conhecido como **alias**. O aliasing acontece quando o valor de uma variável é atribuído a outra variável porque as variáveis são apenas nomes que armazenam referências ao valor real.

```python
import matplotlib.pyplot as plt
```

Agora, o pacote Pyplot pode ser chamado de plt.

Exemplo: (Desenhar uma linha em um diagrama da posição (0,0) até a posição (6,250)) 
> (x,y)

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints,color="r")
plt.show()
```

### $\color{blue}{\text{Plotagem Matplotlib}}$

#### $\color{red}{\text{Plotando pontos x e y}}$

A função **plot()** é usada para desenhar pontos (marcadores) em um diagrama.

Por padrão, a função **plot()** desenha uma linha de ponto a ponto.

A função recebe parâmetros para especificar pontos no diagrama.

O parâmetro 1 é uma matriz contendo os pontos no eixo x .

O parâmetro 2 é uma matriz contendo os pontos no eixo y .

Se precisarmos traçar uma linha de (1, 3) a (8, 10), temos que passar dois arrays [1, 8] e [3, 10] para a função plot.

Exemplo: (Desenhar uma linha em um diagrama da posição (1, 3) para a posição (8, 10))

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, color="y")
plt.show()
```

O **eixo x** é o eixo horizontal.

O **eixo y** é o eixo vertical.

#### $\color{red}{\text{Plotagem sem linha}}$

Para plotar apenas os marcadores, você pode usar o parâmetro de notação de string de atalho 'o', que formará um círculo.

Exemplo: (Desenhar dois pontos no diagrama, um na posição (1, 3) e outro na posição (8, 10))

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

#### $\color{red}{\text{Vários pontos}}$

Exemplo:(Desenhar uma linha em um diagrama da posição (1, 3) para (2, 8), depois para (6, 1) e finalmente para a posição (8, 10))

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints, color="r")
plt.show()
```

#### $\color{red}{\text{Pontos X padrão}}$
Se não especificarmos os pontos no eixo x, eles obterão os valores padrão 0, 1, 2, 3, (etc. dependendo do comprimento dos pontos y.

Então, se pegarmos o mesmo exemplo acima e deixarmos de fora os pontos x, o diagrama ficará assim:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints, color="r")
plt.show()
```

Os pontos x no exemplo acima são [0, 1, 2, 3, 4, 5].

### $\color{blue}{\text{Marcadores Matplotlib}}$

#### $\color{red}{\text{Marcadores}}$
Você pode usar o argumento de palavra-chave **marker** para enfatizar cada ponto com um marcador especificado.

Exemplo: (Marcar cada ponto com um círculo)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

Exemplo: (Marcar cada ponto com uma estrela)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '*')
plt.show()
```

#### $\color{red}{\text{Strings de formato fmt.}}$

Também é possível usar o parâmetro de notação de string de atalho para especificar o marcador.

Este parâmetro também é chamado **fmt**, e é escrito com esta sintaxe:

` marker | line| color`

Exemplo: (Marcar cada ponto com um círculo)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:g')
plt.show()
```

#### $\color{red}{\text{Referência de linha}}$

| Marcador  | Descrição   |
| ------- | -------- |
| '-'   | Linha sólida (Solid)    |
| ':'   | Linha pontilhada (Dotted)   |
| '--'   | Linha traçada (Dashed)   |
| '-.'   | Linha pontinhada/traçada  (Dashed/dotted)  |

>Se deixar de fora o valor da linha no parâmetro fmt, nenhuma linha será plotada.

#### $\color{red}{\text{Tamanho do marcador}}$

O argumento de palavra-chave **markersize** ou a versão mais curta **ms** são usadas para definir o tamanho dos marcadores:

Exemplo: (Definir o tamanho dos marcadores para 20)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o',color="c", ms = 20)
plt.show()
```

#### $\color{red}{\text{Tamanho do marcador}}$

O argumento de palavra-chave **markeredgecolor** ou **mec** para definir a cor da borda dos marcadores.

Exemplo: (Definir a cor em volta do marcador para vermelho)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

Pode também usar o argumento de palavra-chave **markerfacecolor** ou **mfc** para definir a cor dentro da borda dos marcadores.

Exemplo: (Definir a cor da parte de dentro do marcador)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

Use os **mec** e **mfc** para colorir todo o marcador 

Exemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

```python
ypoints = np.array([5, 8, 1, 10])
plt.plot(ypoints, marker = '*', ms = 15,color="k", mec = 'hotpink', mfc = 'hotpink')
plt.show()
```
