### $\color{blue}{\text{Linha Matplotlib}}$
#### $\color{red}{\text{Estilo de linha}}$
Use a palavra-chave argument **linestyle**, ou short **ls**, para alterar o estilo da linha plotada.

Exemplo: (Usar uma linha pontilhada) 

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

Exemplo: (Usar a linha tracejada)

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed') 
plt.show()
```

#### $\color{red}{\text{Sintaxe mais curta}}$

O estilo de linha pode ser escrito em uma sintaxe mais curta:

- linestyle pode ser escrito como `ls`

- dotted pode ser escrito como `:`

- dashed pode ser escrito como `--`

Refazendo os exemplos anteriores:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = '--')
plt.show()
```

#### $\color{red}{\text{Cor da linha}}$
Pode usar o argumento de palavra-chave **color** ou **c** para definir a cor da linha.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'c')
plt.show()
```

#### $\color{red}{\text{Espessura da linha}}$
Pode usar o argumento de palavra-chave **linewidth** ou **lw** para alterar a largura da linha.

Exemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '18.5')
plt.show()
```

#### $\color{red}{\text{Várias linhas}}$
Pode plotar quantas linhas quiser simplesmente adicionando mais funções **plt.plot()**.


Exemplo: (Desenhar duas linhas especificando plt.plot() para cada linha)

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2,color='hotpink')

plt.show()
```

Também é possível plotar muitas linhas adicionando os pontos para os eixos xey para cada linha na mesma função **plt.plot()**.

Exemplo: (Desenhar duas linhas especificando os valores em x e y)

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

### $\color{blue}{\text{Rótulos e títulos}}$


#### $\color{red}{\text{Criar rótulos para um gráfico}}$

Com o Pyplot, você pode usar as funções **xlabel()** e **ylabel()** para definir um rótulo para os eixos x e y.

Exemplo:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y,color="y")

plt.xlabel("Amperes")
plt.ylabel("Resistência")

plt.show()
```

#### $\color{red}{\text{Criar um título para o gráfico}}$

Com o Pyplot, podemos usar a função **title()** para definir um título para o gráfico.

Exemplo: (Adicionar um título de plotagem e rótulos para os eixos x e y)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Tensão")
plt.xlabel("Amperes")
plt.ylabel("Resistência")

plt.show()
```

#### $\color{red}{\text{Definir propriedades de fonte para título e rótulos}}$

Para definir as propriedades da fonte para o título e os rótulos podemos usar o parâmetro **fontdict** em xlabel(), ylabel()e title().

Exemplo: (Definir as propriedades da fonte para o título e rótulos)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Serotonina",fontdict = font1)
plt.xlabel("ME",fontdict = font2)
plt.ylabel("Soshis",fontdict = font2)
plt.plot(x, y)
plt.show()
```

#### $\color{red}{\text{Posicione o título}}$

Podemos usar o  parâmetro **loc** em title() para posicionar o título. Utilizando valores como: 'left','right e 'center'.

Exemplo:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Serotonina",fontdict = font1,loc = 'left')
plt.xlabel("ME",fontdict = font2)
plt.ylabel("Soshis",fontdict = font2)
plt.plot(x, y)
plt.show()
```

### $\color{blue}{\text{Adicionando Linhas de Grade}}$

#### $\color{red}{\text{Adicionar linhas de grade a um gráfico}}$
Com o Pyplot, podemos usar a função **grid()** para adicionar linhas de grade ao gráfico.

Exemplo: (Adicionar linhas de grade ao gráfico)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Data")
plt.xlabel("Study")
plt.ylabel("Orders")

plt.plot(x, y)

plt.grid()

plt.show()
```

#### $\color{red}{\text{Especificar quais linhas de grade exibir}}$

Podemos usar o parâmetro **axis** na função **grid()** para especificar quais linhas de grade exibir.
Os valores legais são: 'x', 'y' e 'both'. O valor padrão é 'both'
Exemplo: (Exibir apenas linhas de grade para o eixo x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Data")
plt.xlabel("Study")
plt.ylabel("Orders")

plt.plot(x, y)

plt.grid(axis='x')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Data")
plt.xlabel("Study")
plt.ylabel("Orders")

plt.plot(x, y)

plt.grid(axis='y')

plt.show()
```

#### $\color{red}{\text{Definir propriedades de linha para a grade}}$

É possível definir as propriedades da linha da grade, como: grid(color = 'color ', linestyle = 'linestyle', linewidth = number ).

Exemplo: (Definir as propriedades da linha da grade)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Data")
plt.xlabel("Study")
plt.ylabel("Orders")

plt.plot(x, y, color='r')

plt.grid(color='c',ls='--',lw=0.7)

plt.show()
```

##### $\color{orange}{\text{Estilo XKCD}}$
O estilo XKCD é um modo de desenho estilo esboço xkcd, tornando o efeito do gráfico como um desenho.
Pra isso usamos **plt.xkcd()** para ativar o estilo antes de definir a função. 
E para desativar basta colocar **plt.rcdefaults()** após a função.

Exemplo:(Plotando um gráfico com XKCD)

```python
import numpy as np
import matplotlib.pyplot as plt
# Ativar o estilo
plt.xkcd()

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2,color='hotpink')

plt.show()
# Desativar o estilo
plt.rcdefaults()
```

```python
with plt.xkcd(): #Exemplo do Matplotlib
    # Based on "Stove Ownership" from XKCD by Randall Munroe
    # https://xkcd.com/418/

    fig = plt.figure()
    ax = fig.add_axes((0.1, 0.2, 0.8, 0.7))
    ax.spines.right.set_color('none')
    ax.spines.top.set_color('none')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_ylim([-30, 10])

    data = np.ones(100)
    data[70:] -= np.arange(30)

    ax.annotate(
        'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
        xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

    ax.plot(data)

    ax.set_xlabel('time')
    ax.set_ylabel('my overall health')
    fig.text(
        0.5, 0.05,
        '"Stove Ownership" from xkcd by Randall Munroe',
        ha='center')
```
