### $\color{blue}{\text{Subplot}}$
#### $\color{red}{\text{Exibir vários gráficos}}$

Com a função **subplot()** podemos desenhar vários gráficos em uma figura.

Exemplo: (Desenhar 2 gráficos)

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #(linhas,colunas,índice do gráfico(1º,2º,3º..))
plt.plot(x,y,color='r')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y,color='c')

plt.show()
```

#### $\color{red}{\text{Função subplot()}}$
A função **subplot()** recebe três argumentos que descrevem o layout da figura.

O layout é organizado em linhas e colunas, que são representadas pelo primeiro e segundo argumento.

O terceiro argumento representa o índice do gráfico atual.

>plt.subplot(1, 2, 1)
a figura tem 1 linha, 2 colunas, e este gráfico é o primeiro gráfico

>plt.subplot(1, 2, 2)
a figura tem 1 linha, 2 colunas, e este gráfico é o segundo gráfico.

Então, se queremos uma figura com 2 linhas e 1 coluna (o que significa que os dois gráficos serão exibidos um em cima do outro em vez de lado a lado), podemos escrever a sintaxe assim:

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1) #(linhas,colunas,índice do gráfico(1º,2º,3º..))
plt.plot(x,y,color='r')

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)#(linhas,colunas,índice do gráfico(1º,2º,3º..))
plt.plot(x,y,color='c')

plt.show()
```

Exemplo: (Desenhar 6 gráficos)

```python
import matplotlib.pyplot as plt
import numpy as np
#plot 1
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y,color='r')
#plot 2
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y,color='k')
#plot 3
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y,color='b')
#plot 4
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y,color='g')

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
#plot 5
plt.subplot(2, 3, 5)
plt.plot(x,y,color='y')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
#plot 6
plt.subplot(2, 3, 6)
plt.plot(x,y,color='c')

plt.show()
```

#### $\color{red}{\text{Título}}$

Com a função **title()** adicionamos o título

Exemplo: (2 gráficos com títulos)

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) #(linhas,colunas,índice do gráfico(1º,2º,3º..))
plt.plot(x,y,color='r')
plt.title("gráfico 1")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y,color='c')
plt.title("gráfico 2")
plt.show()
```

#### $\color{red}{\text{Super Título}}$

Podemos adicionar um título à figura inteira com a função **suptitle()**.

Exemplo: (Adicionar um título para figura inteira)

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1) 
plt.plot(x,y,color='r')
plt.title("gráfico 1")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y,color='c')
plt.title("gráfico 2")

plt.suptitle("Função Suptitle")
plt.show()
```

### $\color{blue}{\text{Scatter/Dispersão}}$
#### $\color{red}{\text{Criando gráficos de dispersão}}$

Com o Pyplot, você pode usar a função **scatter()** para desenhar um gráfico de dispersão.

A função **scatter()** traça um ponto para cada observação. Ele precisa de duas matrizes do mesmo comprimento, uma para os valores do eixo x e outra para os valores do eixo y.

Exemplo: (Fazer um gráfico de dispersão simples)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y,c='c')
plt.show()
```

A observação no exemplo acima é o resultado de 13 carros passando. O eixo X mostra a idade do carro. O eixo Y mostra a velocidade do carro quando ele passa.

#### $\color{red}{\text{Comparar parcelas}}$

É basicamente utilizar dois gráficos de dispersão na mesma figura.

Exemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)

plt.show()
```

Ao comparar os dois gráficos, ambos nos dão a mesma conclusão: quanto mais novo o carro, mais rápido ele dirige.

#### $\color{red}{\text{Colorir cada ponto}}$

Podemos até definir uma cor específica para cada ponto usando uma matriz de `colors` como valor para o argumento **c** .
>Não pode usar o **color** para isso, apenas o **c**

Exemplo: (Definir várias cores dos marcadores)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

#### $\color{red}{\text{Mapa de cores}}$
O módulo Matplotlib tem vários mapas de cores disponíveis.
Um mapa de cores é como uma lista de cores, onde cada cor tem um valor que varia de 0 a 100.
Como este: ![exemplo](https://www.w3schools.com/python/img_colorbar.png)
Este mapa de cores é chamado de 'viridis' e, varia de 0, que é uma cor roxa, e até 100, que é uma cor amarela

##### $\color{orange}{\text{Como usar o ColorMap}}$

Podemos especificar o mapa de cores com o argumento de palavra-chave **cmap** com o valor do mapa de cores, neste caso, 'viridis'que é um dos mapas de cores integrados disponíveis no Matplotlib.

Além disso, deve ser criado um array com valores (de 0 a 100), um valor para cada ponto no gráfico de dispersão.
A inclusão do mapa de cores no desenho se dá a partir declaração **plt.colorbar()**.
Exemplo: (Criar uma matriz de cores e especificar um mapa de cores no gráfico de dispersão)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

#### $\color{red}{\text{Tamanho}}$
Podemos alterar o tamanho dos pontos com o argumento **s** (size).

Assim como as cores, certifique-se de que a matriz para tamanhos tenha o mesmo comprimento que as matrizes para os eixos x e y.

Exemplo: (Definir o tamanho para os marcadores)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

#### $\color{red}{\text{Alfa}}$

Podemos ajustar a transparência dos pontos com o argumento **alpha**.

Assim como as cores, certifique-se de que a matriz para tamanhos tenha o mesmo comprimento que as matrizes para os eixos x e y.

Exemplo:(Definir o tamanho para os marcadores)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

#### $\color{red}{\text{Combinar tamanho de cor e alfa}}$
Podemos combinar um mapa de cores com diferentes tamanhos nos pontos. Isso é melhor visualizado se os pontos forem transparentes.

Exemplo:(Criar matrizes aleatórias com 100 valores para pontos x, pontos y, cores e tamanhos)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```
