### $\color{blue}{\text{Barras}}$
#### $\color{red}{\text{Criando Barras}}$

Com o Pyplot, você pode usar a função **bar()** para desenhar gráficos de barras.

Exemplo: (Desenhar 4 barras em um gráfico)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

A função **bar()** recebe argumentos que descrevem o layout das barras.

As categorias e seus valores representados pelo primeiro e segundo argumento como matrizes.

Exemplo:

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

#### $\color{red}{\text{Barras horizontais}}$
Se você quiser que as barras sejam exibidas horizontalmente em vez de verticalmente, use a função **barh()**.

Exemplo: (Desenhar 4 barras horizontais)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y)
plt.show()
```

#### $\color{red}{\text{Cor da barra}}$
O **bar()** e **barh()** usa o argumento de palavra-chave **color** para definir a cor das barras.

Exemplo: ( Desenhar 4 barras vermelhas)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "r")
plt.show()
```

##### $\color{orange}{\text{Nomes de cores}}$

Podemos usar quais [cores](https://matplotlib.org/stable/gallery/color/named_colors.html) quisermos.

Exemplo: (Desenhar 4 barras rosa choque)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "hotpink")
plt.show()
```

#### $\color{red}{\text{Largura da Barra}}$

O **bar()** usa o argumento de palavra-chave **width** para definir a largura das barras.

Exemplo: (Desenhar 4 barras muito finas)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.1,color="hotpink")
plt.show()
```

O valor de largura padrão é 0,8

>Para barras horizontais, use **height** em vez de **width*.

#### $\color{red}{\text{Altura da Barra}}$

A **barh()** leva o argumento de palavra-chave **height** para definir a altura das barras.

Exemplo: (Desenhar 4 barras finas)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

O valor de altura padrão é 0,8

### $\color{blue}{\text{Histogramas Matplotlib}}$
#### $\color{red}{\text{Histograma}}$

Um histograma é um gráfico que mostra as distribuições de frequência .

É um gráfico que mostra o número de observações dentro de cada intervalo dado.

Exemplo: digamos que você peça a altura de 250 pessoas, você pode acabar com um histograma como: ![exemploo](https://www.w3schools.com/python/img_matplotlib_histogram1.png)

#### $\color{red}{\text{Criar histograma}}$

No Matplotlib, usamos a função  para criar histogramas.

A função **hist()** usará uma matriz de números para criar um histograma, a matriz é enviada para a função como um argumento. Usamos o NumPy para gerar aleatoriamente uma matriz com 250 valores, onde os valores se concentrarão em torno de 170 e o desvio padrão é 10

Exemplo: (Faça uma distribuição de dados normal por NumPy)

```python
import numpy as np

x = np.random.normal(170, 10, 250)

print(x)
```

A função **hist()** lerá o array e produzirá um histograma

Exemplo: (histograma simples)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x, color='g')
plt.show()
```

### $\color{blue}{\text{Gráficos de pizza}}$
#### $\color{red}{\text{Criando gráficos de pizza}}$

Com o Pyplot, usamos a função **pie()** para desenhar gráficos de pizza.

Exemplo: (Gráfico de pizza simples)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

O gráfico de pizza desenha uma peça (chamada de cunha(ou fatia para fazer lógica)) para cada valor na matriz (neste caso [35, 25, 25, 15]).
> A cunha é uma formação gráfica que indica continuação de tendência. É parecida com triângulos simétricos no tempo de sua formação e também pelo fato de os preços ficarem variando entre duas linhas convergentes.

Por padrão, a plotagem da primeira fatia começa no eixo x e se move no sentido anti-horário. Ou seja começa entre o azul e vermelho, em um ângulo de 0° graus.

>O tamanho de cada cunha é determinado comparando o valor com todos os outros valores, usando esta fórmula:

> O valor dividido pela soma de todos os valores: **x/sum(x)**

#### $\color{red}{\text{Rótulos}}$
Para adicionar rótulos ao gráfico de pizza basta usar o parâmetro **label**.

O **label** deve ser uma matriz com um rótulo para cada fatia.

Exemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]

plt.pie(y, labels = m)
plt.show() 
```

#### $\color{red}{\text{Ângulo inicial}}$

Como mencionado, o ângulo inicial padrão está no eixo x, mas podemos alterar o ângulo inicial especificando um parâmetro **startangle**.

O **startangle** é definido com um ângulo em graus, o ângulo padrão é 0.



Exemplo: (Começar a primera cunha a 90 graus)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]

plt.pie(y, labels = m, startangle = 90)
plt.show() 
```

#### $\color{red}{\text{Explode}}$
O parâmetro **explode** permite que uma cunha se destaque.

O **explode ** deve ser uma matriz com um valor para cada fatia.

Cada valor representa a que distância do centro cada cunha é exibida.

Exemplo: (Coloque a cunha uvas em 0,2 do centro da pizza)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]
uva = [0, 0, 0, 0.2]

plt.pie(y, labels = m, explode = uva)
plt.show() 
```

#### $\color{red}{\text{Sombra}}$

Adicione uma sombra ao gráfico de pizza definindo o parâmetro **shadows** como `True`.

Exemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]
uva = [0, 0, 0, 0.2]

plt.pie(y, labels = m, explode = uva, shadow = True)

plt.show() 
```

#### $\color{red}{\text{Cores}}$

O parâmetro **colors**, se especificado, deve ser uma matriz com um valor para cada fatia.

Exemplo: (Especificar uma nova cor para cada fatia)

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]
novas = ["black","c","pink","yellow"]

plt.pie(y, labels = m, colors=novas)
plt.show() 
```

[Mais cores](https://matplotlib.org/stable/gallery/color/named_colors.html).

#### $\color{red}{\text{Legenda}}$
Para adicionar uma lista de explicação para cada cunha, use a função **legend()**.

Exemplo: 

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]
novas = ["black","c","pink","yellow"]

plt.pie(y, labels = m, colors=novas)
plt.legend()
plt.show() 
```

#### $\color{red}{\text{Legenda com cabeçalho}}$
Para adicionar um cabeçalho à legenda, adicione o parâmetro **title** à função **legend** .

Exemplo:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
m = ["Maçãs", "Bananas", "Abacaxis", "Uvas"]
novas = ["black","c","pink","yellow"]

plt.pie(y, labels = m, colors=novas)
plt.legend(title = "Frutas")
plt.show() 
```
