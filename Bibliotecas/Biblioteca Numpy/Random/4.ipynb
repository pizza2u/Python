### $\color{blue}{\text{Distribuição Exponencial}}$

A distribuição exponencial é usada para descrever o tempo até o próximo evento, por exemplo, falha/sucesso, etc. Utilizando o método **.exponential()**.

Tem dois parâmetros:

- scale: inverso da taxa (lam na distribuição de poisson) padrão para 1,0.

- size: A forma da matriz retornada.

Exemplo: (Desenhar uma amostra para distribuição exponencial com escala 2.0 com tamanho 2x3)

```python
from numpy import random

x = random.exponential(scale=2, size=(2, 3))

print(x)
```

#### $\color{red}{\text{Visualização da Distribuição Exponencial}}$

Exemplo:

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.exponential(size=1000),color="r",hist=False)

plt.show()
```

#### $\color{red}{\text{Relação entre Poisson e Distribuição Exponencial}}$

A distribuição de Poisson lida com o número de ocorrências de um evento em um período de tempo, enquanto a distribuição exponencial lida com o tempo entre esses eventos.


### $\color{blue}{\text{Distribuição Qui-quadrado}}$

A distribuição do Qui-quadrado é usada como base para verificar a hipótese. Usando o método **.chisquare()**.
Tem dois parâmetros:

- df: (grau de liberdade).

- size: A forma da matriz retornada.

Exemplo: (Desenhar uma amostra para distribuição qui-quadrado com grau de liberdade 2 com tamanho 2x3)

```python
from numpy import random

x = random.chisquare(df=2, size=(2, 3))

print(x)
```

#### $\color{red}{\text{Visualização da distribuição do Qui-quadrado}}$

Exemplo:

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.chisquare(df=1, size=1000), color="b", hist=False)

plt.show()
```

### $\color{blue}{\text{Distribuição Rayleigh}}$

A distribuição de Rayleigh é usada no processamento de sinais. Utilizando o método **.rayleigh()**.

Tem dois parâmetros:

- scale: (desvio padrão) decide quão plana a distribuição será o padrão 1).

- size: A forma da matriz retornada.

Exemplo: (Extrair uma amostra para distribuição rayleigh com escala de 2 com tamanho 2x3)

```python
from numpy import random

x = random.rayleigh(scale=2, size=(2, 3))

print(x)
```

#### $\color{red}{\text{Visualização da distribuição do Qui-quadrado}}$

Exemplo:



```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.rayleigh(size=1000), hist=False)

plt.show()
```

### $\color{blue}{\text{Distribuição de Pareto}}$
Uma distribuição seguindo a lei de Pareto, ou seja, distribuição 80-20 (20% de fatores causam 80% de resultado). Utilizando o método **.pareto()**.

Tem dois parâmetros:

- a: parâmetro de forma.

- size: A forma da matriz retornada.

Exemplo: (Extrair uma amostra para distribuição de Pareto com forma de 2 com tamanho 2x3)

```python
from numpy import random

x = random.pareto(a=2, size=(2, 3))

print(x)
```

#### $\color{red}{\text{Visualização da distribuição de Pareto}}$

Exemplo:



```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.pareto(a=2, size=1000),color="r", kde=False)

plt.show()
```

### $\color{blue}{\text{Distribuição Zipf}}$
As distribuições Zipf são usadas para amostrar dados com base na lei de Zipf. Utilizando o método **.zipf()**.

>Lei de Zipf: Em uma coleção, o enésimo termo comum é 1/n vezes o termo mais comum. Por exemplo, a 5ª palavra comum em português ocorre quase 1/5 da palavra mais usada.

Tem dois parâmetros:

- a: parâmetro de distribuição.

- size: A forma da matriz retornada.

Exemplo: (Desenhar uma amostra para distribuição zipf com parâmetro de distribuição 2 com tamanho 2x3)

```python
from numpy import random

x = random.zipf(a=2, size=(2, 3))

print(x)
```

#### $\color{red}{\text{Visualização da Distribuição Zipf}}$

Exemplo: (1000 pontos, mas plotando apenas aqueles com valor < 10 para um gráfico mais significativo)


```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

x = random.zipf(a=2, size=1000)
sns.distplot(x[x<10],color="r", kde=False)

plt.show()
```
