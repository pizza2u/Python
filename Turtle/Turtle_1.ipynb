## $\color{blue}{\text{Biblioteca Turtle (from python)}}$

É uma biblioteca Python pré-instalada que permite aos usuários criar imagens e formas, mostrando uma tela virtual.

A caneta na tela que você usa para desenhar é chamada de **turtle** e é isso que dá nome à biblioteca. Em resumo, a biblioteca ajuda os programadores a terem uma ideia de como é a programação com Python de uma maneira divertida e interativa.

Com a biblioteca Python turtle, você pode desenhar e criar vários tipos de formas e imagens. Como por exemplo o símbolo do TikTok:

```python
from turtle import *

#tiktok
width(20)
bgcolor('black')
colors=['#db0f3c','#50ebe7','white']
pos = [(0,0),(-5,13),(-5,5)]

for (x,y),col in zip(pos,colors):
    up()
    goto(x,y)
    down()
    color(col)
    left(180)
    circle(50,270)
    forward(120)
    left(180)
    circle(50,90)

done()
```

#### Para iniciar

O **turtle** uma biblioteca gráfica, o que significa que é preciso uma janela separada para executar cada comando de desenho. Você pode criar esta tela inicializando uma variável para ela.
Em Python, você usa variáveis para armazenar informações que usará posteriormente em seu programa. 
Você inicializa uma variável quando atribui um valor inicial a ela. Como o valor da variável não é constante, ele pode mudar várias vezes durante a execução do seu programa.

##### Exemplo 1 (Inicializando uma variável)

```python
import turtle
s = turtle.getscreen()
```

Esta janela é chamada de tela. É onde você pode visualizar a saída do seu código. A pequena forma triangular preta no meio da tela é chamada de turtle.
Em seguida, você inicializa a variável t(ou qualquer nome que queira dar), que você usará em todo o programa para se referir à turtle:

```python
 t = turtle.Turtle()
```

Agora você tem sua tela e sua turtle. A tela funciona como uma tela (*não diga*), enquanto a turtle funcionará como uma caneta. Você pode programar a turtle para se mover pela tela. Ela tem certas características mutáveis, como tamanho, cor e velocidade. E sempre aponta em uma direção específica e se moverá nessa direção, a menos que você diga o contrário:

- Quando está para cima, significa que nenhuma linha será desenhada quando ele se mover.
- Quando está para baixo, significa que uma linha será desenhada quando se mover.

#### Programando com o Turtle


A primeira coisa que você aprenderá quando se trata de programação com a biblioteca turtle é como fazer a caneta se mover na direção que você deseja. 

Existem quatro direções em que uma turtle pode se mover:

- Para frente
- Para trás
- Direita
- Esquerda

A turtle se move por meio do comandos:
- forward() 
- backward() 

E você pode mudar essa direção girando-a através de:
- left() 
- right()

Em um certo grau. 

##### Exemplo 2 (Movendo a turtle)

```python
t.right(90)
t.forward(100)
t.left(90)
t.backward(100)
```

Quando você executa esses comandos, a turtle vira à direita em 90° graus, e avança 100 unidades, vira à esquerda em 90° graus e recua 100 unidades.

Também há abreviações para esses comandos: 

t.rt() **para** t.right()

t.fd() **para** t.forward()

t.lt() **para** t.left()

t.bk() **para** t.backward()

A tela é dividida em quatro quadrantes. O ponto onde a turtle está inicialmente posicionada no início do seu programa é (0,0). 

Isso é chamado de Home. Para mover a turtle para qualquer outra área da tela, você usa **.goto()**

```python
t.goto(100,100)
```

Para trazer a tartaruga de volta à sua posição inicial usa-se:

```python
t.home()
```
